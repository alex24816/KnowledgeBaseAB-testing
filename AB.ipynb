**Пример** - компания Барака Обамы 2008 года. Выбор между видео и изображением в пользу изображения позволил собрать дополнительные 60 млн долл. пожертвований. 

Если внедряется что-то совершенно новое, то для контрольной группы сделать наивный бейзлайн (просто правила) и использовать его не только для сравнения качества модели, но и для принятия решения о её эффективности в АБ тесте

Две группы
* **Контрольная** - не изменяем ничего
* **Тестовая** - с изменениями

**p-value** - вероятность получить для случайной величины такое же или более экстремальное значение статистики (среднего арифметического, медианы и др.), по сравнению с ранее наблюдаемым, при условии, что нулевая гипотеза верна. 

**Неверное утверждение**
p-value - вероятность, что $H_0$ верна

Статистика - любая (измеримая) функция от выборки.

## <center> A/A-тестирование </center>

A/A тестирование - тестирование, в котором обе группы контрольные. Это позволяет провести эксперимент от начала и до конца и расчитать метрики без внедрения новой функциональности.  
Это надежный способ узнать о статистической значимости эксперимента и протестировать инфраструктуру эксперимента.


## Планирование A/B-теста

Все этапы эксперимента должны быть определены заранее.

Список контрольных вопросов (пока кажется, что только 3 вопросы интересны):

**Цель**
- в чем цель этого теста?  

**Зоны ответственности**
- кто представитель от бизнеса? 
- кто отвечает за реализацию теста?
- кто осуществляет аналитику?

**Планирование эксперимента**
- кто входит в тестовую и контрольную группы?
- каковы нулевая и альтернативная гипотеза?
- какие показатели планируется отслеживать?
- когда начало тестирования?
- **с какого момента начинается эксперимент для аналитических целей (есть ли эффект новизны)?**
- **сколько продлится тест?**
- **каков размер выборки?**
